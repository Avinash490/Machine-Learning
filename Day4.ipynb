```python
# concatenate two data frame 
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([d1,d2])
```

```python
# concatenate two data frame (axis)
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([d1,d2], axis = 1)
```

```python
# concatenate two data frame 
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2],)
```

```python
# concatenate two data frame (outer)
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2], axis = 1, join = "outer")
```

```python
# concatenate two data frame (inner)
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2], axis = 1, join = "inner")
# jo data frame comman ho use hi print kraye ga inner me..
```

```python
# key parameter in pandas 
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
pd.concat([d1,d2], axis = 1, keys=["d1","d2"])
```

```python
# key parameter in pandas without axis..
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([d1,d2],  keys=["d1","d2"])
```

```python
# key parameter in pandas without axis..
import pandas as pd 
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"C":[1,2,3,4],"D":[21,22,23,24]})
pd.concat([d1,d2],  keys=["d1","d2"])
# col. same n ho to bo apna col. bna lega..
```

```python
#Group by in pandas 
import pandas as pd 
d1 = pd.DataFrame({"Name":["a","b","c","d","a","b","c"],"s1":[12,13,14,12,13,14,15],"s2":[12,13,14,12,13,14,15]})
d2 = d1.groupby("Name")
for x,y in d2:
    print (x) 
    print (y)
```

```python
#pivot function is used to reshape
import pandas as pd 
d1 = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","a","b","a","b"],"eng":[10,12,14,16,18,10],"math":[12,13,114,115,17,18]})
d1.pivot(index = "days",columns = "st_name")
```
