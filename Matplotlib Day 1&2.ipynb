```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,1,3,5,3,6,4]
c = [10, 20, 30, 40, 50, 60, 70]
s = [100,150,200,250,300,350,400]
plt.xlabel("language")
plt.ylabel("city")
plt.title("language and its city")
scatter = plt.scatter(x,y, c = c, cmap = "viridis" , marker = "*")
colorbar = plt.colorbar(scatter)
colorbar.set_label("colorbar",fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,4,5,6]
area2 = [3,2,5,6,1]
area3 = [2,4,1,4,2]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3, labels = l)
plt.title("mera h")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc= 2)
# legend fun is used to denotes the color of the region.
# It takes location as arguments.
plt.show()
```

```python
# Matplotlib is used for making graphs, charts etc
import matplotlib.pyplot as plt
# In this matplotlib is a library and pyplot a class of matplotlib.
x = [1,2,3,4]
y = [2,5,11,0.4]
c = ["red","green","black","blue"]
size = [100,200,300,400]
# this is line used to resize the pointer in the graphs
# For this matplotlib has already a predefined var "s".
# Edge color is used to make the outer line of the pointer.

plt.scatter(x,y , color = c,marker = "*",s = size,edgecolor = "y")
# Here marker is used to change the pointer in the charts.
plt.show()
```

```python
# Matplotlib is used for making graphs, charts etc
import matplotlib.pyplot as plt
# In this matplotlib is a library and pyplot a class of matplotlib.
x = [1,2,3,4]
y = [2,4,6,8]

plt.bar(x,y)

plt.show()

# Here show() fun is used to display the graphs.

# Here are going to design the graphs

c = ["red","green","yellow","blue"]
# We can also use this.
# c = ["r","g","y","b"]
plt.bar(x,y,color= c)
plt.show()

# Here we are going to give the label(title) to the graphs
# In this label we can change the font size of the labels by using fontsize keyword.
plt.xlabel("baby", fontsize = 30)
plt.ylabel("bachha")
plt.title("Matplotlib")
plt.bar(x,y, color= c)
plt.show()
```
